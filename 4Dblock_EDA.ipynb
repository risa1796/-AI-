```python
import pandas as pd

train_df = pd.read_csv('./train.csv')
test_df = pd.read_csv('./test.csv')
```

```python
train_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(3, 3, figsize=(20, 15))
axes = axes.flatten()

for ax,col in zip(axes,train_df[list(train_df.iloc[:,2:].columns)]):
  print(train_df[col].value_counts())
  train_df[col].value_counts().plot(kind='bar',ax=ax, color=['green', 'red'])

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
from glob import glob
import os 
import numpy as np

image_paths = glob(os.path.join('./train/',  '*'))

for _ in range(5):
  idx = np.random.randint(0,len(image_paths))
  image_name = image_paths[idx]
  image = Image.open(image_name)
  print(train_df[train_df['id'] == image_name[-15:-4]].iloc[:,2:])
  
  plt.title(image_name[-15:-4])
  plt.imshow(image)
  plt.show()

  print('')
  print('')
```
